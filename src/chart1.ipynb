```python
res=requests.get('https://dailyview.tw/top100/topic/9')
soup=BeautifulSoup(res.text,'html.parser')

figures = []
divs=soup.find_all("div", ["volume-rank-row"])
divs=divs[:10]
for div in divs:
    figures.append([s for s in div.stripped_strings])

names, goods, ordis, bads = [], [], [], []
for i in range (10):
  names.append(figures[i][1])
  goods.append((int)(figures[i][6]))
  ordis.append((int)(figures[i][9]))
  bads.append((int)(figures[i][12]))
```

```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.io as pio
pio.renderers.default = 'colab'
from plotly.offline import init_notebook_mode, iplot
```

```python
# Guage chart https://plotly.com/python/v3/gauge-charts/
for k in range(10):
    def transform_xy(xy):
        if xy is 30:
            x=str(0.24)
            y=str(0.6531)
            return (str)(x+' '+y)
        elif xy < 30:
            denominator=(xy-10)/50
            x=(str)(0.24-0.06*(math.cos(math.pi*denominator)))
            y=(str)(0.5+0.1531*(math.sin(math.pi*denominator)))
            return (str)(x+' '+y)
        elif xy>30:
            denominator=(60-xy)/50
            x=(str)(0.24+0.06*(math.cos(math.pi*denominator)))
            y=(str)(0.5+0.1531*(math.sin(math.pi*denominator)))
            return (str)(x+' '+y)
    a=transform_xy(100 * goods[k]/(bads[k]+goods[k])) # a: approval rating

    base_chart= {
    "values": [40, 10, 10, 10, 10, 10, 10],
    "labels": [" ","10", "20", "30", "40", "50", "60"],
    "domain": {"x": [0, .48]},
    "marker": {
        "colors": [
            'rgb(255, 255, 255)',
            'rgb(255, 255, 255)',
            'rgb(255, 255, 255)',
            'rgb(255, 255, 255)',
            'rgb(255, 255, 255)',
            'rgb(255, 255, 255)',
            'rgb(255, 255, 255)'
        ],
        "line": {"width": 1}
    },
    "name": "Gauge",
    "hole": .4,
    "type": "pie",
    "direction": "clockwise",
    "rotation": 108,
    "showlegend": False,
    "hoverinfo": "none",
    "textinfo": "label",
    "textposition": "outside"
    }


    meter_chart = {
        "values": [50, 10, 10, 10, 10, 10],
        "labels": ["Approval rating", "超母湯", "普通母湯", "母湯", "微母湯", "支持度過半"],
        "marker": {
            'colors': [
                'rgb(255,255,255)',
                'rgb(153,77,82)',
                'rgb(217,116,43)',
                'rgb(230,180,80)',
                'rgb(138,151,123)',
                'rgb(182, 194, 154)'
            ]
        },
        "domain": {"x": [0, 0.48]},
        "name": "Gauge",
        "hole": .3,
        "type": "pie",
        "direction": "clockwise",
        "rotation": 90,
        "showlegend": False,
        "textinfo": "label",
        "textposition": "inside",
        "hoverinfo": "none"
    }

    layout = {
        'xaxis': {
            'showticklabels': False,
            'showgrid': False,
            'zeroline': False,
        },
        'yaxis': {
            'showticklabels': False,
            'showgrid': False,
            'zeroline': False,
        },
        'shapes': [
            {'type':'path',
             'path': 'M 0.235 0.5 L' + a +'L 0.245 0.5 Z',
             'fillcolor': 'rgba(1,77,103,1)',
             'line':{'width':0.5},
             'xref':'paper',
             'yref':'paper'}
        ],
    

          
        'annotations': [{'xref': 'paper','yref': 'paper','x': 0.22,'y': 0.45,'text': (str)(names[k]),'showarrow': False}],

    }
    base_chart['marker']['line']['width'] = 0



    fig = {"data": [base_chart, meter_chart],
       "layout": layout}
    
    py.iplot(fig, filename=drive_path+str(names[k]))
    #py.plot(fig, drive_path+str(names[k])) # comment out this line to save the figure as html
```
